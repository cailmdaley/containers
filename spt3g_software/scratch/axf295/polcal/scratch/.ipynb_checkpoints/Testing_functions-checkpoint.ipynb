```python
def calc_source_masked_variance(bolo_map,Noise_Mask=''):

    '''
    Calculate the variance of a masked map
    Checks the shapes of both maps and does the right thing.
    
    Arguments
    ---------
    bolo_map : 2D numpy array
        map to be masked and to of which to find the variance
        
    noise_mask : 2D numpy array
        mask map; multiplies bolo_map to mask source.
    
    Returns
    -------
    variance: float
        variance of the masked map (where bolo_map*noise_mask !=0)
        
    '''

    if Noise_Mask == '':
        Noise_Mask = np.zeros(np.shape(bolo_map))
        Noise_Mask[np.where(abs(bolo_map)>1.)]+=1.
        Noise_Mask = 1.0-Noise_Mask
    ms = np.shape(Noise_Mask)[0]
    dsx = np.shape(bolo_map)[0]
    dsy = np.shape(bolo_map)[1]
    if dsx > ms or dsy > ms:
        bolo_map = bolo_map[dsx//2-ms//2:dsx//2+ms//2, dsy//2-ms//2:dsy//2+ms//2]
        bolonoisemask = Noise_Mask
    elif dsx < ms or dsy < ms:
        bolonoisemask = Noise_Mask[ms//2-dsx//2:ms//2+dsx//2, ms//2-dsy//2:ms//2+dsy//2]
    else:
        bolonoisemask = Noise_Mask
        
    #plt.imshow(bolonoisemask);plt.show()
    #plt.imshow(bolonoisemask*bolo_map);plt.colorbar();plt.show()
    try:
        variance = np.nanvar(bolo_map[np.nonzero(bolonoisemask)])
    except IndexError:
        variance = -1
    return variance

mapvar_perband = {}
obss = glob.glob('/big_scratch/axf295/2019/CenA/Analyzed_Data/7*')
obsids = []
for ob in obss:
    obsids.append(ob.split('/')[-1])
    
for obs in sorted(obsids):
    try:
        nomangcoadds = pk.load(open('/big_scratch/axf295/2019/CenA/Analyzed_Data/%s/%s_NomAngCoaddMaps.pkl'%(obs,obs),'rb'))
    except Exception:
        continue
    allcoadds = nomangcoadds['all']
    mapvar_perband[obs] = {}
    plt.figure()
    for ang in sorted(allcoadds):
        for band in allcoadds[ang]:
            #plt.imshow(allcoadds[ang][band]);plt.colorbar()
            #plt.show()
            mapvar = np.var(allcoadds[ang][band][60:,:20])#calc_source_masked_variance(allcoadds[ang][band])#
            if band not in mapvar_perband[obs]:
                mapvar_perband[obs][band] = []
            mapvar_perband[obs][band].append(mapvar)
            #print(ang,band,mapvar)
            plt.scatter(ang,mapvar,color=BANDCOLORS[band])
            #imdat = np.copy(allcoadds[ang][band])
            #imdat[abs(imdat)>1.] = 0.
            #plt.imshow(imdat);plt.colorbar()
            #print(np.nanvar(imdat))
    plt.grid()
    plt.xlabel('ang')
    plt.ylabel('mapvar')
    plt.title(obs)
    plt.show()
    for band in mapvar_perband[obs]:
        print(band,np.mean(mapvar_perband[obs][band]),np.std(mapvar_perband[obs][band]))
    print('---------------------------')
```

```python
import Circle_Functions as CF
def get_optimal_nomang_rotation(nomangfits):
    ## minimize chi^2 by rotating nominal angles
    ## fit to 90GHz since those are the highest SN
    nomangs = sorted(nomangfits.keys())
    fits = []
    stds = []
    for ang in nomangs:
        fits.append(nomangfits[ang]['90']['rawPhi'][0])
        stds.append(nomangfits[ang]['90']['rawPhi'][1])
    print(fits,nomangs)
    var = np.var(np.array(fits)-nomangs)
    dof = len(nomangs)
    rotangle = np.linspace(0,180.,181)
    chi2 = []
    plt.errorbar(nomangs,fits,yerr=stds);plt.plot([0,180],[0,180],color='k',ls='--')
    plt.title('no rotation');plt.show()
    for i in range(len(rotangle)):
        angs = np.copy(nomangs)-rotangle[i]
        #print(angs)
        angs[angs>180.]-=180.
        angs[angs<0.]+=180.
        chi2.append(np.nansum((CF.difference(angs,fits,deg=True))**2/stds)/(dof-1))
        
        #plt.errorbar(nomangs,bestfits,yerr=stds);plt.plot([0,180],[0,180],color='k',ls='--')
        #plt.title(rotangle[i]);plt.show()
        
    plt.plot(rotangle,chi2);plt.show()
    bestang = rotangle[np.where(chi2==min(chi2))]
    bestfits = fits-bestang
    bestfits[bestfits<0.]+=180
    bestfits[bestfits>180.]-=180
    print(bestfits)
    print(bestang)
    
    plt.errorbar(nomangs,bestfits,yerr=stds);plt.plot([0,180],[0,180],color='k',ls='--');plt.show()
    
    return chi2


nomangfits = pk.load(open('/big_scratch/axf295/2019/CenA/Analyzed_Data/77216509/77216509_NomAngCoadd_BestFitParams.pkl','rb'))

chi2 = get_optimal_nomang_rotation(nomangfits['all'])
```
